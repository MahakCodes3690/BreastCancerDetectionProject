```python
df.drop(columns=['Unnamed: 32'],inplace=True,axis=1)
df.drop(columns=['id'],inplace=True,axis=1)
```

```python
x = df.iloc[:, 1:]
y = df.iloc[:, 0]
```

```python
df['diagnosis'].value_counts()
```

### **Train test split and Handle Class Imbalance with SMOTE**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=21)
```

# **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
st_x=StandardScaler()
x_train=st_x.fit_transform(x_train)
x_test=st_x.transform(x_test)
```

#  Model Selection & Fitting

### **Support vector machine**

```python
from sklearn.svm import SVC

svm_model=SVC(kernel='linear')

# model fitting
svm_model.fit(x_train,y_train)

# model prediction
y_pred1=svm_model.predict(x_test)
```

## **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

logistic_model=LogisticRegression()

# model fitting
logistic_model.fit(x_train,y_train)

# model prediction
y_pred2=logistic_model.predict(x_test)
```

## **Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier

decision_model=DecisionTreeClassifier(max_depth=6, min_samples_split=5,random_state=42)

# model fitting
decision_model.fit(x_train,y_train)

#model prediction
y_pred3=decision_model.predict(x_test)
```

## **Random forest**

```python
from sklearn.ensemble import RandomForestClassifier

random_forest_model=RandomForestClassifier(n_estimators=200,random_state=42)

# model fitting
random_forest_model.fit(x_train,y_train)

#model prediction
y_pred4= random_forest_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('SVM Model Accuracy:',accuracy_score(y_test,y_pred1))
print('Logistic Regression Model Accuracy:',accuracy_score(y_test,y_pred2))
print('Decision Tree Model Accuracy',accuracy_score(y_test,y_pred3))
print('Random Forest Model Accuracy',accuracy_score(y_test,y_pred4))
```

After evaluating four machine learning models  Support Vector Machine (SVM), Logistic Regression, Decision Tree, and Random Forest  the SVM model achieved the best overall performance. The SVM obtained an accuracy of 99.12%, outperforming the Decision Tree (94.73%) and Random Forest (95.61%) models. Logistic Regression also achieved 99.12% accuracy; however, SVM demonstrated slightly better generalization and stability on unseen data.

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print('SVM Classification Report:',classification_report(y_test,y_pred1))
```

```python
print('SVM Confusion matrix:',confusion_matrix(y_test,y_pred1))
```

In summary, the Support Vector Machine (SVM) model performed exceptionally well, achieving a high overall accuracy of 99% with balanced precision and recall across both benign and malignant classes.  The confusion matrix revealed that the SVM correctly classified nearly all samples, with only one false prediction. This indicates that the SVM model is highly reliable and effective for distinguishing between cancerous and non-cancerous cases, making it an ideal choice for breast cancer detection and diagnosis.

```python
import joblib
from sklearn.pipeline import Pipeline
# Build pipeline: standard scaler + SVM classifier (probability True if you want probabilities)
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("svc", SVC(kernel="linear", probability=True, random_state=42))
])

pipeline.fit(x_train, y_train)

# Save pipeline
joblib.dump(pipeline, "breast_cancer_pipeline.joblib")
print("Saved breast_cancer_pipeline.joblib")
```
